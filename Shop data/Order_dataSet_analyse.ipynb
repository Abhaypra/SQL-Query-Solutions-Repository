# **Import necessary Libraries**








```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

```

# **Read CSV file**

```python
data=pd.read_csv('/content/orders.csv',na_values=['Not Available', 'unknown'])
data.head(10)
```

# **Data Cleaning**

```python
data['Ship Mode'].unique()
```

```python
data.columns=data.columns.str.lower()
data.columns=data.columns.str.replace(' ','_')
```

```python
data.isnull().sum()
data.dropna(inplace=True)
data.isnull().sum()
```

```python
data.duplicated().sum()
```

```python
data.head()
```

```python
data.describe()
```

# **Feature Engineering**

```python
data['discount_value']=data['list_price']*(data['discount_percent']/100)
data['sale_price']=data['list_price']-data['discount_value']
data['profit']=data['sale_price']-data['cost_price']
data.head()
```

```python
# data=data.drop(columns=['cost_price','list_price','discount_percent'])
```

```python
from datetime import datetime
data.dtypes
data['order_date']=pd.to_datetime(data['order_date'],format='%Y-%m-%d')
data.dtypes
```
